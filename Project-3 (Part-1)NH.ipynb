```python
### 3.Univariate Analysis ###
```

```python
# Visualize the Distribution
# Histogram
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(Ds1['SalePrice'], bins=30, kde=False, color='blue')
plt.title('Histogram of House Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

```python
# Kernel Density Estimate (KDE) Plot
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(Ds1['SalePrice'], color='red', fill=True)
plt.title('KDE Plot of House Prices')
plt.xlabel('Price')
plt.ylabel('Density')
plt.grid()
plt.show()
```

```python
#Combined Histogram and KDE Plot
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(Ds1['SalePrice'], bins=30, kde=True, color='blue', stat='density', alpha=0.5)
plt.title('Histogram and KDE of House Prices')
plt.xlabel('Price')
plt.ylabel('Density')
plt.grid(axis='y')
plt.show()
```

```python
# Summary statistics
```

```python
summary_stats = Ds1['SalePrice'].describe()
print(summary_stats)
```

```python
# Identify Outliers
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=Ds1['SalePrice'], color='lightgreen')
plt.title('Box Plot of House Prices')
plt.xlabel('Price')
plt.grid(axis='x')
plt.show()
```

```python
#Data Preprocessing
```

```python
# Check for missing values
print(Ds1.isnull().sum())

# Fill or drop missing values as necessary
Ds1.fillna(method='ffill', inplace=True)  # Example: forward fill

# Convert categorical variables to numerical if necessary
Ds1 = pd.get_dummies(Ds1, drop_first=True)  # One-hot encoding
```

```python
# Correlation Matrix
```

```python
# Calculate the correlation matrix
correlation_matrix = Ds1.corr()

# Set up the matplotlib fgure
plt.figure(figsize=(12, 10))

# Draw the heatmap
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})

# Title and show the plot
plt.title('Correlation Matrix of House Prices Dataset')
plt.show()

sns.heatmap(Ds1.corr(numeric_only=True), annot=True, cmap='coolwarm')
```

```python
# Correlation heatmap
corr_matrix = Ds1[['SalePrice', 'BedroomAbvGr', 'FullBath', 'BsmtUnfSF', 'LotArea', 'MSSubClass']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Feature Correlation with Price')
plt.show()
```

```python
#Pairplot (Scatterplot Matrix)
```

```python
# Select a subset of features for the pairplot
features = ['OverallQual', 'GrLivArea', 'GarageCars', 'TotalBsmtSF', 'SalePrice']  # Example features
sns.pairplot(Ds1[features], diag_kind='kde')

# Show the plot
plt.suptitle('Pairplot of Selected Features', y=1.02)
plt.show()
```

```python
# Scatter Plots
```

```python
# Scatter plot for GrLivArea vs SalePrice
plt.figure(figsize=(10, 6))
sns.scatterplot(data=Ds1, x='GrLivArea', y='SalePrice')
plt.title('GrLivArea vs SalePrice')
plt.xlabel('Above Ground Living Area (sq ft)')
plt.ylabel('Sale Price ($)')
plt.show()
```

```python
# Scatter plot for OverallQual vs SalePrice
plt.figure(figsize=(10, 6))
sns.boxplot(data=Ds1, x='OverallQual', y='SalePrice')
plt.title('Overall Quality vs SalePrice')
plt.xlabel('Overall Quality Rating')
plt.ylabel('Sale Price ($)')
plt.show()
```
